# matplotlib简单用法
--------------


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline      #图片内联
```

```python
#主图画法
x=np.linspace(-np.pi,np.pi,256,endpoint=True)   #linspace函数第三参数是点个数
s,c=np.sin(x),np.cos(x)
plt.plot(x,s)
plt.plot(x,c)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)       #设置宽高比（10:6），其中宽为80
plt.plot(x,s,color="blue",linewidth=2.5,linestyle="-")
plt.plot(x,c,color="red",linewidth=2.5,linestyle="-")       #设置线颜色，线宽，样式
```

```python
#改变坐标轴刻度
x=np.linspace(-np.pi,np.pi,256,endpoint=True)   #linspace函数第三参数是点个数
s,c=np.sin(x),np.cos(x)
plt.plot(x,s)
plt.plot(x,c)
plt.xlim(x.min()*2,x.max()*2)
plt.ylim(s.min()*2,s.max()*2)
plt.ylim(c.min()*2,c.max()*2)
```

```python
#增加图例
plt.plot(x,s,color="blue",linewidth=2.5,linestyle="-",label="sin")
plt.plot(x,c,color="orange",linewidth=2.5,linestyle="-",label="cos")
plt.legend(loc="upper left",frameon=False)
```
